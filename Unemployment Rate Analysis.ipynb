```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\BAIBHAV RAJ MAHARANA\Desktop\Oasis Infobyte\Unemployment in India.csv')
```

```python
# Assuming you have already loaded the DataFrame 'df'

# Convert the 'Region' column to string type
df['Region'] = df['Region'].astype(str)

# Convert the ' Estimated Unemployment Rate (%)' column to float type
df[' Estimated Unemployment Rate (%)'] = df[' Estimated Unemployment Rate (%)'].astype(float)

# Now create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['Region'], df[' Estimated Unemployment Rate (%)'])
plt.title('Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)  # Rotates x-axis labels for better visibility
plt.tight_layout()  # Adjusts the layout for better spacing
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\BAIBHAV RAJ MAHARANA\Desktop\Oasis Infobyte\Unemployment in India.csv')
```

```python


# Converting the 'Region' column to string type
df['Region'] = df['Region'].astype(str)

# Convert the ' Estimated Unemployment Rate (%)' column to float type
df[' Estimated Unemployment Rate (%)'] = df[' Estimated Unemployment Rate (%)'].astype(float)

# Now creating a bar graph
plt.figure(figsize=(18, 16))
plt.bar(df['Region'], df[' Estimated Unemployment Rate (%)'])
plt.title('Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)  # Rotates x-axis labels for better visibility
plt.tight_layout()  # Adjusts the layout for better spacing
plt.show()
```

```python
#INSIGHTS Of the Unemployment Rate Analysis :-
#1 Puducherry has the highest unemployment rate of about 76.74 percent of total population.
#2 Assam has the lowest unemployment rate of about 11.17 percent of total population.
#3 The unemployment rate of all the states has Drastically increased from 2019 to 2020 and we all know,
# the reason behind this surprising drastic increase in unemployment is due to the Covid-19 Pandemic,
# which ultimately became the reason of Global Recession.
```
